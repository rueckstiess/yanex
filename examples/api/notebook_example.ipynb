```python
import yanex.results as yr

# get completed experiments with a limit of 10
experiments = yr.get_experiments(limit=10, status="completed")

# print the results in a readable format
for experiment in experiments:
    print(f"Experiment ID: {experiment.id}")
    print(f"Name: {experiment.name}")
    print(f"Description: {experiment.description}")
    print(f"Status: {experiment.status}")
    print(f"Started At: {experiment.started_at}")
    print(f"Completed At: {experiment.completed_at}")
    print(f"Params: {experiment.get_params()}")
    print(f"Metrics: {experiment.get_metrics()}")
    print("-" * 40)
```
